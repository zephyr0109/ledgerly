# 자산 보고서 생성 (모듈화 버전)

```python
from ledgerly.reports import (
    generate_asset_report_content, 
    save_markdown, 
    export_to_pdf, 
    OUTPUT_DIR
)
```

```python
report_date = "2026-02-24"
print("보고서 생성 중...")
markdown_content = generate_asset_report_content(report_date)
```

```python
base_name = f"asset_report_{report_date[:7]}"
md_path = OUTPUT_DIR / f"{base_name}.md"
pdf_path = OUTPUT_DIR / f"{base_name}.pdf"

save_markdown(markdown_content, md_path)
success = export_to_pdf(markdown_content, pdf_path)

if success:
    print(f"Markdown: {md_path}")
    print(f"PDF: {pdf_path}")
    print("Markdown 및 PDF 저장 완료")
else:
    print("PDF 변환 실패. Markdown만 저장되었습니다.")
```
