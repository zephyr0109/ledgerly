```python
import pandas as pd
import sqlite3
from ledgerly.utils import find_project_root

# DB 연결
conn = sqlite3.connect(find_project_root() /"data" /"db"/"ledgerly.db")

# CSV 읽기
income_df = pd.read_csv(
    find_project_root() / "data" / "raw" / "income_form.csv",
    dtype={
        "일자": "string",
        "금액": "int64",
        "수입구분": "string",
        "비고": "string",
    }
)

# 컬럼 매핑
mapped_income_df = pd.DataFrame()
mapped_income_df["income_date"] = income_df["일자"]
mapped_income_df["amount"] = income_df["금액"]
mapped_income_df["income_type"] = income_df["수입구분"]
mapped_income_df["memo"] = income_df["비고"]

# DB 저장
mapped_income_df.to_sql(
    "income",
    conn,
    if_exists="append",
    index=False
)

conn.close()
```
